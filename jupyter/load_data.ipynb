```python
# 从第一次筛选得到的知识点序列
def get_skill(user):

    skill=pd.read_csv('../result/assistment2009/assist2009_recommend_skill.csv',sep=',')
    skill=skill.fillna(-1).to_numpy()
    skill=skill[user]
    skill_0=[]
    for i in skill:
        if i!=-1:
            skill_0.append(int(i))
    # print(skill)
    return skill_0
skill=get_skill(0)
print(len(skill))
```

```python
# 找到该学生对应该知识点所对应的备选习题集合
######有几个限制条件，学生没有做过的题目，题目不重复，是一个集合
def get_negative_problem(user):
    exercise=pd.read_csv('../data/assistment2009/assisment2009_apriori_input.csv',sep=',')
    exercise=exercise.drop_duplicates(['user_id','problem_id'],keep='first')
    exercise_old=exercise.loc[exercise['user_id']==user].to_numpy()
    old=[]
    for i in exercise_old:
        old.append(i[1])
    # print(old)

    exercise=exercise.to_numpy()
    # print(exercise)
    skill_set=get_skill(user)
    # print(skill_set)
    exercise_set=[]
    exercise0=[]
    for x in skill_set:

        for i in exercise:
            if i[2]==x:
                # print(i)
                exercise_set.append(i[1])
        
    exercise0.append(set(exercise_set))
    # print(exercise0)
    exercise_all=[]
    a=[]
    for i in exercise0[0]:
        if i in exercise_03_05:
            # print(i)
            a.append(i)
            if i not in old:
                exercise_all.append(i)
    # print("len(a):",len(a))
    # print(a)
    # print("len(exercise_all",len(exercise_all))
    # print("exercise_all:",exercise_all)
    return exercise_all

# negative=[]
# for i in range(0,4217):
#     negative.append(problem)
# negative=pd.DataFrame(negative)
# negative.to_csv(r'C:\Users\wangxin\Desktop\remmend_DKT_Apriori_DMF\code\DMF\data\assistment2009.test.negative',index=False)
```

```python
def get_negative_skill_exercise(user,skill):
    exercise=pd.read_csv('../data/assistment2009/assisment2009_apriori_input.csv',sep=',')
    exercise=exercise.drop_duplicates(['user_id','problem_id'],keep='first')
    exercise_old=exercise.loc[exercise['user_id']==user].to_numpy()
    old=[]
    for i in exercise_old:
        old.append(i[1])
    # print(old)

    exercise=exercise.to_numpy()
    # print(exercise)
    # skill_set=get_skill(user)
    # print(skill_set)
    exercise_set=[]
    exercise0=[]
    for i in exercise:
        if i[2]==skill:
            # print(i)
            exercise_set.append(i[1])
        
    exercise0.append(set(exercise_set))
    # print(exercise0)
    exercise_all=[]
    a=[]
    for i in exercise0[0]:
        if i in exercise_03_05:
            # print(i)
            a.append(i)
            if i not in old:
                exercise_all.append(i)
    # print("len(a):",len(a))
    # print(a)
    # print("len(exercise_all",len(exercise_all))
    # print("exercise_all:",exercise_all)
    return exercise_all
```

```python
# 从DMF中得到排序后的候选习题
def get_DMF_sort_exercise(user):
    with open('../result/assistment2009/DMF_03_05_exercise.csv') as f:
        exercises=f.readlines()
        first_exercise=exercises[2*(user+1)-1].split(' ')
        first_exercise=[int(i) for i in first_exercise]
    return first_exercise
    # with open('../result/assistment2009/DMF_03_05_index.csv') as f:
    #     indexs=f.readlines()
    #     first_index=indexs[199].split(' ')
    #     first_index=[float(i) for i in first_index]
    # # print(len(first_exercise))
    # print(len(first_index))
    # DMF_x=first_exercise
    # DMF_y=first_index
```

```python
#得到推荐结果(每个知识点推荐五个题目，TOP_5)
def get_recommend_result(user):
    houxuan_exercise=get_negative_problem(user)
    houxuan_skill=get_skill(user)
    # print("len(skill)",len(houxuan_skill))
    DMF_sort_exercise=get_DMF_sort_exercise(user)
    recommend_result=[]
    for i in houxuan_skill:
        recommend_exercise=get_negative_skill_exercise(user,i)
        k=1  #控制每个知识点推荐的个数
        # print(i)
        for j in DMF_sort_exercise:
            
            if k<=5:#top_k  K=5
                if j in recommend_exercise:
                    
                    k=k+1
                    recommend_result.append(j)
                    # print(k)
                
    return recommend_result

# 需要跑34个小时   
recommend_result_all=[]
for i in tqdm.tqdm(range(0,4098)):
    recommend_result_all.append(get_recommend_result(i))
recommend_result_all

```
