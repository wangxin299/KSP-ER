```python
# 自编码器
AUTO=pd.read_csv('../result/assistment2009/AUTO_99.csv',sep=',',names=[ 'item', 'rating'])
# print(AUTO)

df_data_order0 = AUTO.sort_values(by=['rating'],ascending=[False])
# print(df_data_order0)
AUTO99_x=df_data_order0['item'].to_numpy()
AUTO99_y=df_data_order0['rating'].to_numpy()
# d=df_data_order0['rating'].head(20).to_numpy()
len(AUTO99_x)
```

```python
# 对于DMF模型的结果进行处理**********

with open('../result/assistment2009/DMF_03_05_exercise.csv') as f:
    exercises=f.readlines()
    first_exercise=exercises[1].split(' ')
    first_exercise=[int(i) for i in first_exercise]

with open('../result/assistment2009/DMF_03_05_index.csv') as f:
    indexs=f.readlines()
    first_index=indexs[1].split(' ')
    first_index=[float(i) for i in first_index]
print(len(first_exercise))
print(len(first_index))
DMF_x=first_exercise
DMF_y=first_index
```
